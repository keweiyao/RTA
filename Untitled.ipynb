```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1)
for t0 in [0.001, 0.01, 0.1, 1, 10]:
    t, y1, y2 = np.loadtxt("./tau0={}.dat".format(t0)).T
    ax.plot(t, y1, color='orange')
    ax.plot(t, y2, color='blue')
ax.set_xlabel(r"$\tau/\tau_R$")
ax.set_ylabel(r"$[30]/[10]$")
ax.semilogx()
ax.set_ylim(-4,0.1)
plt.savefig("check-the-same-initial-condition.png",dpi=300)
```
